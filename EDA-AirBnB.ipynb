```python

# Convert bedrooms to numeric
df["bedrooms"] = pd.to_numeric(df["bedrooms"], errors="coerce")

# Convert baths to numeric (remove text like 'Half-bath')
df["baths"] = df["baths"].astype(str).str.extract(r"(\d+\.?\d*)").astype(float)

# Convert rating to numeric if it’s a score
df["rating"] = df["rating"].astype(str).str.extract(r"(\d+\.?\d*)").astype(float)

# Convert last_review to datetime
df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")
```

```python
df.dtypes
```

```python
num_cols = [
    "price", "minimum_nights", "number_of_reviews", "reviews_per_month",
    "availability_365", "beds", "bedrooms", "baths",
    "calculated_host_listings_count", "rating"
]

for col in num_cols:
    if col in df.columns:
        plt.figure()
        sns.histplot(df[col], bins=50, kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()

        # Boxplot to check outliers
        plt.figure()
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()
```

```python
cat_cols = [
    "neighbourhood_group", "neighbourhood", "room_type",
    "host_name", "license"
]

for col in cat_cols:
    if col in df.columns:
        plt.figure()
        top_vals = df[col].value_counts().head(10)  # Top 10 categories
        sns.barplot(x=top_vals.values, y=top_vals.index)
        plt.title(f"Top categories in {col}")
        plt.xlabel("Count")
        plt.ylabel(col)
        plt.show()
```
